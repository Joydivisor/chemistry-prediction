### 步骤 1：理解问题

**什么是溶解度？**

溶解度（Solubility）是指物质在溶剂中溶解的能力。在药物开发中，水溶解度（logS）是一个关键指标：
- logS 值越高，药物越容易溶解在水中
- 溶解度影响药物的吸收效率和生物利用度
- 溶解度太低的药物难以被人体吸收

**我们的7个特征及其物理意义**：

1. **MolWt（分子量）**：分子的质量，通常分子量越大，溶解度越低
2. **LogP（脂水分配系数）**：衡量分子的疏水性，LogP 越高越疏水，溶解度越低
3. **NumHDonors（氢键供体数量）**：能提供氢键的原子数，增加溶解度
4. **NumHAcceptors（氢键受体数量）**：能接受氢键的原子数，增加溶解度
5. **TPSA（拓扑极性表面积）**：分子表面的极性区域面积，影响溶解度
6. **NumRotatableBonds（可旋转键数量）**：分子的柔性，影响溶解度
7. **AromaticProportion（芳香性比例）**：芳香环的比例，影响溶解度

**任务目标**：
使用这7个特征，预测分子的水溶解度（logS）。

## 常见问题和调试

### 问题 1：训练时出现 ConvergenceWarning

**原因**：模型在最大迭代次数内未收敛

**解决方案**：
- 增加 `max_iter` 参数（如 2000 或 5000）
- 调整学习率 `learning_rate_init`
- 检查特征是否标准化
- 尝试不同的优化器 `solver`（'adam', 'sgd', 'lbfgs'）

### 问题 2：训练集准确率很高，测试集很低

**原因**：过拟合

**解决方案**：
- 增加正则化参数 `alpha`（如从 0.0001 增加到 0.01）
- 减少网络复杂度（更少的层或神经元）
- 使用 `early_stopping=True`
- 增加训练数据
- 使用 Dropout（需要用 Keras/PyTorch）

### 问题 3：训练很慢

**原因**：数据量大或网络复杂

**解决方案**：
- 减少 `max_iter`
- 使用更小的网络
- 使用 'adam' 优化器（通常比 'lbfgs' 快）
- 减少训练数据（用于快速实验）

### 问题 4：不同运行结果不一致

**原因**：随机初始化

**解决方案**：
- 设置 `random_state` 参数
- 多次运行取平均
- 使用交叉验证评估稳定性

### 问题 5：如何选择网络结构？

**经验法则**：
- 输入特征少（<10）：单层 10-50 个神经元
- 输入特征中等（10-100）：单层 50-200 个神经元，或两层
- 输入特征多（>100）：两层或三层，逐层递减
- 先简单后复杂，避免一开始就用很深的网络
