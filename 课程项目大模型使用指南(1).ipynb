# 人工智能导论大模型使用指南
上海大学 钱伟长学院 熊杰

## 一、目的

本说明旨在帮助同学们正确使用大语言模型（LLM）辅助课程学习、期末复习与建模训练。LLM 可以作为学习与思考工具，但不能替代学生本人完成独立分析、建模设计与结果验证。

本课程鼓励同学们将 LLM 用于：

- 理解题意
- 拆解复杂问题
- 组织建模思路
- 比较不同方法
- 检查方案漏洞
- 改进表达与汇报结构

本课程不鼓励将 LLM 用作：

- 直接生成未经理解和验证的最终答案
- 伪造实验、结果或推导过程
- 回避独立思考与方法选择

---

## 二、什么是合适的 LLM 提示词

所谓“合适的提示词”，是指能够让模型清楚理解任务目标、背景信息、输出要求和约束条件的指令。

一个较好的提示词通常应包括以下四个部分：

1. **任务目标**：你希望模型完成什么工作
2. **背景信息**：课程、题目、数据、对象或场景是什么
3. **约束要求**：是否需要简洁、严谨、面向初学者、避免复杂推导等
4. **输出格式**：希望模型按什么结构输出

### 示例

不推荐：

> 帮我做一下这道题。

推荐：

> 请作为建模助教，帮我拆解这道题目。请从“问题目标、已知条件、未知量、约束条件、可选建模方法”五个方面进行分析，要求语言清晰、结构完整、适合本科课程作业使用。

## 三、LLM 在期末复习与考试准备中的合理用途

### 1. 用于理解概念

同学们可以借助 LLM 理解课程中的基本概念、方法之间的联系，以及不同模型的适用条件。

例如：

```text
请用本科生能理解的语言解释什么是过拟合，并说明它为什么会影响模型泛化能力。
```

### 2. 用于拆解题目

对于较复杂的理工科题目或建模题，LLM 可以帮助识别：

- 问题目标
- 已知量与未知量
- 约束条件
- 可行的求解路径

例如：

```text
请帮我拆解下面这道题目，识别其中的目标、已知条件、未知量、约束和关键变量关系。
```

### 3. 用于设计建模流程

在机器学习、数据分析和建模任务中，LLM 可以帮助同学们建立完整的流程意识，例如：

- 数据理解
- 数据预处理
- 特征工程
- 模型选择
- 评估方式
- 误差分析

例如：

```text
请为一个表格数据二分类任务设计课程作业级别的建模流程，要求包含预处理、特征工程、模型选择、评估指标和误差分析。
```

### 4. 用于比较方法

LLM 可以帮助比较不同方法的优缺点、适用场景和实现难度。

例如：

```text
请比较逻辑回归、随机森林和 XGBoost 在中小规模表格分类任务中的适用条件、优点、局限和调参难度。
```

### 5. 用于检查方案问题

LLM 可以作为“审查工具”，帮助发现：

- 数据泄漏
- 不合理的训练/验证划分
- 指标选择不当
- 过拟合风险
- 结论表达过度

例如：

```text
请以课程助教的角度检查下面的建模方案，重点指出数据泄漏、评估失真和过拟合风险。
```

## 四、LLM 如何辅助机器学习建模

在机器学习任务中，LLM 最适合作为“建模助手”而不是“最终决策者”。

### 1. 明确任务类型

首先应确认问题属于：

- 分类
- 回归
- 聚类
- 排序
- 预测

提示词示例：

```text
请判断这个问题更适合定义为分类、回归还是排序任务，并说明原因。
```

### 2. 明确数据特点

在建模前，需要分析：

- 数据类型（表格/图像/文本/时间序列）
- 样本量大小
- 类别是否平衡
- 是否存在缺失值或异常值
- 是否可能存在分布偏移

提示词示例：

```text
请从建模角度分析这个数据集可能存在的主要问题，包括缺失值、类别不平衡、异常值和数据泄漏风险。
```

### 3. 设计完整建模流程

LLM 可以帮助生成结构化流程，例如：

1. 数据清洗
2. 特征处理
3. 划分训练/验证集
4. 基线模型建立
5. 模型比较
6. 调参与验证
7. 结果分析

### 4. 选择合理评估指标

不同任务应采用不同指标。例如：

- 分类：Accuracy、Precision、Recall、F1、AUC
- 回归：MAE、MSE、RMSE
- 类别不平衡时：不宜只看 Accuracy

提示词示例：

```text
当前任务类别不平衡明显，请说明为什么不能只用准确率作为主指标，并推荐更合适的评估指标。
```

### 5. 做误差分析

LLM 可以帮助同学们从以下角度分析模型问题：

- 哪些样本最容易错
- 哪些类别最容易混淆
- 是否对特定条件不鲁棒
- 是否出现过拟合
- 是否存在训练测试分布差异

## 五、LLM 如何辅助理工科问题拆解与建模

理工科中的复杂问题，通常需要先拆解，再建模。LLM 在这个过程中可发挥较大作用。

### 1. 将自然语言题目转为结构化问题

可先识别：

- 研究对象是什么
- 最终目标是什么
- 已知条件有哪些
- 未知量是什么
- 有哪些约束关系

### 2. 判断适合的建模类型

一个理工科问题可能适合：

- 数学公式建模
- 统计建模
- 机理建模
- 数据驱动建模
- 优化建模
- 仿真建模

提示词示例：

```text
请判断下面这个问题更适合使用机理模型、统计模型还是数据驱动模型，并解释原因。
```

### 3. 提出合理假设

建模时通常需要进行适度简化。LLM 可以帮助提出常见假设，例如：

- 忽略次要影响因素
- 假设系统稳态
- 假设噪声服从某种分布
- 假设变量之间局部线性

提示词示例：

```text
请为这个建模问题列出合理的简化假设，并说明这些假设会如何影响结果的准确性和适用范围。
```

### 4. 比较不同建模路径

同一个问题可能有多条解决路线。LLM 可以帮助比较：

- 精度
- 可解释性
- 所需数据量
- 实现难度
- 计算成本

## 六、推荐提示词模板

### 模板 1：概念理解

```text
请作为本课程学习助手，用清晰、规范、适合本科生的语言解释以下概念，并给出一个简单例子。最后总结它与相近概念的区别。
```

### 模板 2：题目拆解

```text
请帮我拆解下面这道题目，按以下结构输出：
1. 问题目标
2. 已知条件
3. 未知量
4. 约束条件
5. 关键变量关系
6. 可选建模方法
7. 推荐方法与理由
```

### 模板 3：机器学习建模

```text
请作为机器学习课程助教，帮我为下面这个任务设计建模方案。请按“问题定义、数据特点、预处理、特征工程、模型选择、评估指标、误差分析、潜在风险”八部分输出。
```

### 模板 4：方案检查

```text
请作为严格的课程助教，检查下面的建模方案。重点检查：数据泄漏、指标不当、过拟合风险、验证方式是否合理、结论是否可信。最后给出修改建议。
```

## 七、使用 LLM 时应注意的事项

### 1. 不能盲目照抄

LLM 给出的内容必须经过学生本人理解、判断与验证。直接照抄会导致：

- 方法与题目不匹配
- 概念使用错误
- 推理链条不完整
- 结果经不起检查

### 2. 要重视验证

特别是在机器学习和建模任务中，任何方案都需要通过：

- 数据实验
- 指标验证
- 对照分析
- 误差分析

来检验其有效性。

### 3. 要主动追问

如果回答过于空泛，可以继续追问：

- 请讲得更具体一些
- 请给出适合本科生实现的版本
- 请说明这种方法的局限
- 请举一个小例子
- 请用表格重新整理

### 4. 要区分“辅助思考”和“替代思考”

课程鼓励同学们使用 LLM 提高学习效率，但不鼓励将其作为替代独立思考的工具。

---

## 八、考试和作业中的基本原则

建议同学们在使用 LLM 时遵守以下原则：

1. **可以用于理解、整理、比较和检查**
2. **不能代替本人完成未经理解的最终答案**
3. **必须对输出内容负责**
4. **必须结合课程知识进行验证和修正**

若课程对 LLM 使用范围另有具体要求，应以课程正式说明为准。

---

## 九、总结

一个合适的 LLM 提示词，核心是把以下几点说明白：

- 任务是什么
- 背景是什么
- 输出要什么形式
- 有什么限制和要求

在机器学习中，LLM 主要可以帮助：

- 理清任务定义
- 设计建模流程
- 比较模型方法
- 选择评估指标
- 进行误差分析

在理工科问题中，LLM 主要可以帮助：

- 拆解题目
- 识别变量与约束
- 提出建模假设
- 比较不同建模路径

正确使用 LLM 的关键，不是让它替你完成任务，而是借助它帮助你更系统、更清楚、更高效地思考问题。
